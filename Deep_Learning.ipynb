```
x= d.iloc[:,1:7]
y = d['Regime1']
```

```
x
```

```
y
```

```
#Data Partition
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 55)     #this ensures the train dataset s the same throughout
yt = y_test
```

```
x_train
```

```
#Normalize the data
mean = x_train.mean(axis = 0)   #This ensures we calculate mean along the columns
x_train -= mean
s = x_train.std(axis = 0)
x_train /= s

x_test -= mean      #We normalize the test data using the mean and standard deviation of the train data
x_test /= s
```

```
#One-Hot Encoding
from keras.utils import to_categorical
y_train = to_categorical(y_train) 
y_test = to_categorical(y_test)  
```

```
y_test
```

```
#Deep Network Model
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout

m = Sequential() 
m.add(Dense(units =32, activation = 'relu', input_dim = 6))   #We have 6 independent variables
m.add(Dropout(0.5)) 
m.add(Dense(units = 16, activation = 'relu')) 
m.add(Dropout(0.5)) 
m.add(Dense(units = 3, activation = 'softmax'))               #The outcome variable has three possible outcomes 
```

```
#Compile the model
m.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy']) 
```

```
#Fit Model
!pip install livelossplot
```

```
#Fit Model
from livelossplot.keras import PlotLossesCallback as p
m.fit(x_train, y_train, batch_size = 1, epochs = 50, validation_split = 0.2, callbacks = [p()]) 
```

```
#Prediction
y_pred = m.predict_classes(x_test) 
y_pred
```

```
m.evaluate(x_test, y_test) 
```

```
#Confusion Matrix
pd.crosstab(y_pred, yt, rownames = ['Predicted'], colnames = ['Actual'], margins = True, margins_name = 'Total') 
```
